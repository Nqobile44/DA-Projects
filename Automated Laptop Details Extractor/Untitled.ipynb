```python
ram_list = []
for index in range(df['Product_Name'].count()):
    if 'RAM' in df.iloc[index]['Product_Name']:
        try:
            deep_index = df.iloc[index]['Product_Name'].split().index("RAM")
        except:
            ram_list.append(df.iloc[index]['Product_Name'])
        else:
            splitted_value = df.iloc[index]['Product_Name'].split()[deep_index-1]
            if '-' in splitted_value:
                ram_list.append(df.iloc[index]['Product_Name'].split()[deep_index+1])

            elif splitted_value.upper() == 'GB':
                ram_list.append(df.iloc[index]['Product_Name'].split()[deep_index-2]+"GB")
            elif '/' in splitted_value:
                ram_list.append(df.iloc[index]['Product_Name'].split()[deep_index-1].split('/')[0])
            else:
                ram_list.append(splitted_value)
    else:
        product = df.iloc[index]['Product_Name'].split()
        package = []
        for item in product:
            if "GB" in item or 'Gb' in item or 'gb' in item:
                if item == "GB":
                    package.append(product[product.index('GB')-1] + product[product.index('GB')])
                else:
                    package.append(item)

                
        if package:
            ram_list.append(package[0])
        else:
            ram_list.append(None)
ram = pd.Series(ram_list)
```

```python
df.insert(loc=1, column='RAM', value=ram)
```

```python
import re
index_list = []
cleaned_index = []
for index in range(180):
    row = df['Product_Name'].iloc[index].upper()
    index = re.search('GB', row, re.IGNORECASE)
    # print(index)
    # print(index.start())
    index_list.append(index)
print(index_list.__len__())
for func in index_list:
    if func:
        cleaned_index.append(func.start())
    else:
        cleaned_index.append(None)

print(cleaned_index)
```

```python
row_index = 0
toxic_product_name = []
for index in cleaned_index:
    if index:
        each_name = df['Product_Name'][row_index][:index-2]
        if each_name == "New ":
            toxic_product_name.append(df['Product_Name'][row_index])
        else:
            toxic_product_name.append(each_name)
    else:
        toxic_product_name.append(df['Product_Name'][row_index])
    row_index += 1
```

```python
sorted_ssd = []
for index in range(180):
    item = df['SSD'][index]
    if item:
        if item.__len__() > 6:
            if '/' in item:
                item = item.split('/')[1]
                sorted_ssd.append(item)
            elif "SSD" in item:
                item = item[3:]
                sorted_ssd.append(item)
            elif '|' in item:
                item = item.split('|')[0]
                sorted_ssd.append(item)
            elif '+' in item:
                item = item.split('+')[1]
                sorted_ssd.append(item)
            elif '-' in item:
                item = item.split('-')[1]
                sorted_ssd.append(item)
        else:
            sorted_ssd.append(item)
    else:
        sorted_ssd.append(item)

final_ssd = []
for value in sorted_ssd:
    if value:
        if value.__len__() == 3 and 'TB' not in value:
            final_ssd.append(np.nan)
        else:
            final_ssd.append(value)
    else:
        final_ssd.append(value)

print(final_ssd.__len__())
```
